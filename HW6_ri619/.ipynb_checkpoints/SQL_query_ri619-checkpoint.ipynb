##### This notebook is set up to link as a default to the fb55 account.  To turn in the homework use the citibike dataset - check that you are able to read in about 40k rows

```python
SQL_SOURCE = 'https://fb55.carto.com/api/v2/sql?q='

import urllib2
import urllib
import StringIO
import ast
import pandas as pd

try:
    from StringIO import BytesIO as io
except ImportError:
    from io import BytesIO as io

def queryCartoDB(query, format='CSV', source=SQL_SOURCE):
    '''queries carto datasets from a given carto account
    Arguments: 
    query - string: a valid sql query string
    format - outlut format  OPTIONAL (default CSV)
    source - a valid sql api endpoint OPTIONAL (default carto fb55 account)
    Returns:
    the return of the sql query AS A STRING
    NOTES:
    designed for the carto API, tested only with CSV return format'''
    
    data = urllib.urlencode({'format': format, 'q': query})
    try:
        response = urllib2.urlopen(source, data)
    except urllib2.HTTPError, e:
        raise ValueError('\n'.join(ast.literal_eval(e.readline())['error']))
    except Exception:
        raise
    return response.read()
```

## Task 2

Select trips only on 2-2-15

Select trips only on the weekend

    What is the average trip duration on the weekend
    
Do the same for weekday

```python
# Select trips only on February 2, 2015
query = '''SELECT * FROM fb55.citibike
WHERE EXTRACT(MONTH FROM starttime::date) IN (02)
AND EXTRACT(DAY FROM starttime::date) IN (02);'''
table = get_data(query)
print("All rides starting on February 2, 2015")
table.head()
```

# Task 3

Show list of start station locations using GROUP BY

Showing number of rides started per station

    But only for stations within 500m of Times Square (40.7577, -73.9857)
    


```python
## Show start station location with GROUP BY
query = '''SELECT start_station_id, start_station_longitude, start_station_latitude FROM fb55.citibike
GROUP BY start_station_id, start_station_longitude, start_station_latitude;'''
table = get_data(query)
print("All start stations with their coordinates")
table.head()
```

```python
## Show start station location with GROUP BY, ordered by number of rides
query = '''SELECT start_station_id, start_station_longitude, start_station_latitude, COUNT(bikeid) as rides
FROM fb55.citibike
GROUP BY start_station_id, start_station_longitude, start_station_latitude
ORDER BY rides DESC;'''
table = get_data(query)
print("All start stations with their coordinates, ordered by the number of rides starting from there")
table.head()
```

```python
## Show start station location with GROUP BY, ordered by number of rides, but only within 500m of Times Square
query = '''SELECT start_station_id, start_station_longitude, start_station_latitude, COUNT(bikeid) as rides
FROM fb55.citibike
WHERE (start_station_longitude > -73.9857 - 0.0045 AND start_station_longitude < -73.9857 + 0.0045) AND
(start_station_latitude > 40.7577 - 0.0045 AND start_station_latitude < 40.7577 + 0.0045)
GROUP BY start_station_id, start_station_longitude, start_station_latitude
ORDER BY rides DESC;'''
table = get_data(query)
print("All start stations with their coordinates, ordered by the number of rides starting from there")
table
```

## Task 4

Find the station that had the longest average trip duration for a station within 500 meters of Times Square

```python
## Find the station that had the longest average trip duration for a station within 500 meters of Times Square
query = '''SELECT start_station_id, start_station_longitude, start_station_latitude, 
AVG(tripduration) as average_trip_duration FROM fb55.citibike
WHERE (start_station_longitude > -73.9857 - 0.0045 AND start_station_longitude < -73.9857 + 0.0045) AND
(start_station_latitude > 40.7577 - 0.0045 AND start_station_latitude < 40.7577 + 0.0045) AND
EXTRACT(DOW FROM starttime::date) IN (0,6)
GROUP BY start_station_id, start_station_longitude, start_station_latitude
ORDER BY average_trip_duration DESC
LIMIT 1;'''
table = get_data(query)
print("Station with the longest average trip duration for a station within 500 meters of Times Square")
table
```
